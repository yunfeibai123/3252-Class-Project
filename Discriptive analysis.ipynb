```python
# Import Libraries
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.api as sm
from IPython.display import Image
from sklearn import linear_model, datasets,metrics
from sklearn.model_selection import train_test_split
import scipy.stats as stats

df=pd.read_csv('export.csv')
```

```python
fig=df.hist(bins=50, figsize=(20,15))
# fig.savefig('hist.png', dpi=125)
```

```python
corr_matrix = df.corr()
```

```python
plt.style.use('seaborn-white')
fig,ax=plt.subplots(1,1,figsize=(15,15))
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr_matrix,annot=True, mask=mask, ax= ax)
fig.savefig('heatmap.png')
```

```python
k = 10
cols = corr_matrix.nlargest(k , 'target')['target'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale = 1.00)
hm = sns.clustermap(cm ,cbar = True,square = True,
                 yticklabels = cols.values, xticklabels = cols.values)
hm.savefig('cluster.png')
```
